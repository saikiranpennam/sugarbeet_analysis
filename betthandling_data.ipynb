```python
def plot_df(df2, x, y, title="", xlabel='Time', ylabel='Clarifier_Level', dpi=100):
    plt.figure(figsize=(15,4), dpi=dpi)
    plt.plot(x, y, color='blue')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()


plot_df(df2, x=df2['Time'], y=df2['Clarifier_Level'], title='Trend and Seasonality')
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from dateutil.parser import parse

#Multiplicative Decomposition
multiplicative_decomposition = seasonal_decompose(df2['Clarifier_Level'], model='multiplicative', period=30)

#Additive Decomposition
additive_decomposition = seasonal_decompose(df2['Clarifier_Level'], model='additive', period=30)

#Plot
plt.rcParams.update({'figure.figsize':(16, 12)})
multiplicative_decomposition.plot().suptitle('Multiplicative Decomposition', fontsize = 16)
plt.tight_layout(rect = [0, 0.03, 1, 0.95])

additive_decomposition.plot().suptitle('Additive Decomposition', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

plt.show()
```

```python
series = df2.loc[:, 'Clarifier_Level'].values
```

```python
#ADF Test - Augmented Dickey Fuller Test

from statsmodels.tsa.stattools import adfuller
result = adfuller(series, autolag='AIC')
print(f'ADF Statistic: {result[0]}')
print(f'n_lags: {result[1]}')
print(f'p-value: {result[1]}')
for key, value in result[4].items():
    print('Critical Values:')
    print(f'    {key}, {value}')
```

```python
#Test for Seasonality
from pandas.plotting import autocorrelation_plot

#Draw Plot
plt.rcParams.update({'figure.figsize':(10, 6), 'figure.dpi':80})
autocorrelation_plot(df2['Clarifier_Level']).tolist()
```

```python
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Draw Plot
fig, axes = plt.subplots(1,2,figsize=(16,3), dpi= 100)
plot_acf(df2['Clarifier_Level'].tolist(), lags=50, ax=axes[0])
plot_pacf(df2['Clarifier_Level'].tolist(), lags=50, ax=axes[1])
```

```python
#Lag Plot

from pandas.plotting import lag_plot
plt.rcParams.update({'ytick.left' : False, 'axes.titlepad':10})

# Plot
fig, axes = plt.subplots(1, 4, figsize=(10,3), sharex=True, sharey=True, dpi=100)
for i, ax in enumerate(axes.flatten()[:4]):
    lag_plot(df2['Clarifier_Level'], lag=i+1, ax=ax, c='firebrick')
    ax.set_title('Lag ' + str(i+1))

fig.suptitle('Lag Plots of Clarifier Level', y=1.05)
plt.show()
```
