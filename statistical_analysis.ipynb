```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load your dataset (replace 'data.csv' with your data file)
df = pd.read_csv('/Users/saikiranpennam/Documents/RA/Western_Sugar_AI_Assessment_Working/2023/Data/Book1.csv')

# Select the three features of interest
selected_features = ['Filter#6_Inlet_Flow', 'Filter#6 _Inlet_Pressure', 'Filter#6_Turbidity']

# Calculate the mean and standard deviation for each feature
feature_stats = df[selected_features].agg([np.mean, np.std])

# Create a grouped bar plot
plt.figure(figsize=(10, 6))
x = np.arange(len(selected_features))  # X-axis positions for bars
bar_width = 0.35  # Width of the bars
opacity = 0.7

# Plot mean values
plt.bar(x, feature_stats.loc['mean'], bar_width, alpha=opacity, label='Mean', color='b', align='center')

# Plot standard deviation values
plt.bar(x + bar_width, feature_stats.loc['std'], bar_width, alpha=opacity, label='Std Dev', color='g', align='center')

# Customize the plot
plt.xlabel('Features')
plt.ylabel('Values')
plt.title('Mean and Standard Deviation for Features')
plt.xticks(x + bar_width / 2, selected_features)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```

```python
# Set the date column as the DataFrame index
df.set_index('DateTime', inplace=True)
```

```python
# Select the three features of interest
selected_features = ['Filter#5_Inlet_Flow', 'Filter#5_Inlet_Pressure', 'Filter#5 _Turbidity']

# Resample the data at 15-minute intervals and calculate the mean
resampled_data = df[selected_features].resample('15T').mean()

# Define unique colors for each feature
colors = ['b', 'g', 'r']

# Create individual subplots for each feature with different colors
plt.figure(figsize=(12, 8))

for i, (feature, color) in enumerate(zip(selected_features, colors), 1):
    plt.subplot(3, 1, i)  # 3 rows, 1 column, i-th subplot
    plt.plot(resampled_data.index, resampled_data[feature], color=color)
    plt.title(feature)
    plt.xlabel('Time')
    plt.ylabel('Mean Value')
    plt.grid(True)

# Adjust subplot layout
plt.tight_layout()

# Show the plot
plt.show()
```
