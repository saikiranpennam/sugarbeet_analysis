#Filter1

```python
#Slicing operation, selecting columns related to FILTER#1 only
filter1_df = df.iloc[3:,1:6]

#Renaming Columns
filter1_df.columns = ['Filter#1CTRL_FRC', 'Filter#1CTRL_PRC', 'Filter#1_InletFlow','Filter#1_InletPressure','Filter#1_Turbidity']
filter1_df
```

```python
#filter1_df.isnull().values.any() --> returns Boolean Value

#Chekcing for NaN values
filter1_df.isna().sum().sum()
```

```python
#Instead of dropping, replacing the Nan with 0
filter1_df = filter1_df.replace(to_replace = np.NaN, value = 0)
filter1_df
```

```python
filter1_df.isna().sum().sum()
```

```python
# Calculating corr between the columns of Filter#1
filter1_df.corr(method = 'pearson', min_periods=1)
```

```python
import seaborn as sns
cormat = filter1_df.corr()
round(cormat, 2)
sns.heatmap(cormat,annot=True, cmap="Blues", linewidths=.7)
```
